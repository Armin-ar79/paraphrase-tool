```python
# ✅ نصب کتابخانه‌ها
!pip install -q transformers gradio

# ✅ وارد کردن ابزارها
from transformers import pipeline
import gradio as gr

# ✅ لود مدل‌ها از Hugging Face
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# ✅ تابع پردازش متن
def analyze_text(text):
    # خلاصه‌سازی
    summary = summarizer(text, max_length=100, min_length=25, do_sample=False)[0]['summary_text']

    # دسته‌بندی موضوعی (با لیست پیش‌فرض)
    labels = ["technology", "health", "education", "science", "sports", "business"]
    classification = classifier(text, candidate_labels=labels)
    top_label = classification["labels"][0]
    score = classification["scores"][0]

    return summary, f"{top_label} ({score:.2f})"

# ✅ رابط Gradio
demo = gr.Interface(
    fn=analyze_text,
    inputs=gr.Textbox(lines=10, placeholder="Paste your text here..."),
    outputs=["text", "text"],
    title="🧠 AI Text Analyzer",
    description="Summarize and classify your text using free Hugging Face models.",
)

# ✅ اجرا
demo.launch()
```
